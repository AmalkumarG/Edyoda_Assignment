```python
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv("train.csv")
test=pd.read_csv("test.csv")
```

```python
train.shape,test.shape
```

```python
train.info()
```

### Get categorical and numerical values from train set and test set 

```python
categorical_train=train.select_dtypes(include=object)
categorical_test=test.select_dtypes(include=object)
```

```python
numerical_train=train.select_dtypes(include=["int64","float64"])
numerical_test=test.select_dtypes(include=["int64","float64"])
```

### Remove features having more than 50% null values

```python
more_than_50_percent=categorical_train.isnull().sum()/len(categorical_train)*100
print(more_than_50_percent[more_than_50_percent>50])
more_than_50_percent=more_than_50_percent[more_than_50_percent>50].index
```

```python
categorical_train=categorical_train.drop(more_than_50_percent,axis=1)
categorical_test=categorical_test.drop(more_than_50_percent,axis=1)
```

### Fill null values using SimpleImputer
Fill categorical values using most_frequent
Fill numerical values using their mean

```python
from sklearn.impute import SimpleImputer
mode_imputer=SimpleImputer(strategy="most_frequent")
median_imputer=SimpleImputer(strategy="median")
```

```python
y_train=numerical_train["SalePrice"]
numerical_train.drop(["SalePrice","Id"],axis=1,inplace=True)
numerical_test.drop(["Id"],axis=1,inplace=True)
```

```python
categorical_train[categorical_train.columns]=mode_imputer.fit_transform(categorical_train)
categorical_test[categorical_test.columns]=mode_imputer.fit_transform(categorical_test)
numerical_train[numerical_train.columns]=median_imputer.fit_transform(numerical_train)
numerical_test[numerical_test.columns]=median_imputer.fit_transform(numerical_test)
```

```python
categorical_train
```

```python
numerical_train
```

### Encode Categorical values of test and train set using get_dummies 

```python
def encode_columns(cat):
    cat1=pd.DataFrame()
    for i in cat:
        dummies=pd.get_dummies(categorical_train[i],drop_first=True)
        cat1=pd.concat([cat1,dummies],axis=1)
    return cat1
```

```python
cat_train=encode_columns(categorical_train.columns)
```

```python
cat_test=encode_columns(categorical_test.columns)
```

### Normalizing test and train numerical values using MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
scaled_train=numerical_train.copy()
scaled_test=numerical_test.copy()
scaler=MinMaxScaler()
```

```python
scaled_train[scaled_train.columns]=scaler.fit_transform(numerical_train)
scaled_test[scaled_test.columns]=scaler.fit_transform(numerical_test)
```

```python
scaled_train
```

```python
scaled_test
```

### Concatinating  numerical dataset and categorical

```python
final_train=pd.concat([scaled_train,cat_train],axis=1)
final_test=pd.concat([scaled_test,cat_test],axis=1)
```

```python
final_train
```

```python
final_test.dropna()
```

### Using Linear regression model for prediction

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(final_train,y_train)
```

```python
reg.score(final_train,y_train)
```
