```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy="most_frequent")
df1=data.copy()
df1[["Union","JF"]]=imputer.fit_transform(df1[["Union","JF"]])
```

```python
len(df1.OG.unique())
```

```python
df2=df1.drop(["OG","Dept","Union","Job"],axis=1)
```

```python
df2
```

```python
df1.DC.value_counts()
```

```python
len(df2.JF.unique())
```

```python
dummies=pd.get_dummies(df2[["JF","YT","DC"]])
```

```python
dummies
```

```python
df3=pd.concat([df2.drop(["JF","YT","DC"],axis=1),dummies.drop(["JF_Supervisory-Labor & Trade","YT_Fiscal","DC_LLB"],axis=1)],axis=1)
```

```python
df3
```

```python
x=df3.drop("Total_Compensation",axis=1)
y=df3.Total_Compensation
```

```python
x
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x[["Year","OGC","UC","EI","Salaries","Overtime","H/D"]]=scaler.fit_transform(x[["Year","OGC","UC","EI","Salaries","Overtime","H/D"]])
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
reg.score(x_test,y_test)
```
