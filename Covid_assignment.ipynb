# 4. Filter the dataframe with only this columns['continent','location','date','total_cases','total_deaths','gdp_per_capita',human_development_index'] and update the data frame.

```python
covid=covid[['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'] ]
```

```python
covid.columns
```

# 5. Data Cleaning
 a. Remove all duplicates observations
 b. Find missing values in all columns
 c. Remove all observations where continent column value is 
missing
 Tip : using subset parameter in dropna
 d. Fill all missing values with 0

```python
covid=covid.drop_duplicates(keep="first")
```

```python
covid.isnull().sum()
```

```python
covid.dropna(subset=["continent"],inplace=True)
```

```python
covid.isnull().sum()
```

```python
covid.fillna(0,inplace=True)
```

```python
covid.isnull().sum()
```

# 7. Data Aggregation:
 a. Find max value in all columns using groupby function on 
'continent'
 column
 Tip: use reset_index() after applying groupby
 b. Store the result in a new dataframe named 'df_groupby'.
 (Use df_groupby dataframe for all further analysis)


```python
covid.groupby("continent").max().reset_index()
```

```python
df_groupby=covid.groupby("continent").max().reset_index()
```

```python
df_groupby
```

#  (Use df_groupby dataframe for all further analysis)

# 8. Feature Engineering :
 a. Create a new feature 'total_deaths_to_total_cases' by ratio of
 'total_deaths' column to 'total_cases'


```python
df_groupby['total_deaths_to_total_cases']=df_groupby.total_deaths/df_groupby.total_cases
```
